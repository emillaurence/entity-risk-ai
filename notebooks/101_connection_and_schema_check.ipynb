# 01 — Connection & Schema Check
Verify Neo4j connectivity and inspect the graph schema.

```python
import sys
sys.path.insert(0, "..")
```

```python
from src.config import get_neo4j_settings
from src.storage.neo4j_repository import Neo4jRepository

settings = get_neo4j_settings()
print(settings.masked())
```

## Connect & test

```python
repo = Neo4jRepository(**vars(settings))

repo.test_connection()
print("Connected to Neo4j")
```

## Node labels

```python
labels = repo.get_labels()
print(f"Labels ({len(labels)}):")
for label in labels:
    print(f"  {label}")
```

## Relationship types

```python
rel_types = repo.get_relationship_types()
print(f"Relationship types ({len(rel_types)}):")
for rel in rel_types:
    print(f"  {rel}")
```

## Property keys

```python
prop_keys = repo.get_property_keys()
print(f"Property keys ({len(prop_keys)}):")
for key in prop_keys:
    print(f"  {key}")
```

## Node counts by label

```python
import pandas as pd

node_counts = repo.get_node_counts_by_label()
df = pd.DataFrame(node_counts.items(), columns=["Label", "Count"])
df = df.sort_values("Count", ascending=False).reset_index(drop=True)
display(df)
```

## Relationship counts by type

```python
rel_counts = repo.get_relationship_counts_by_type()
df = pd.DataFrame(rel_counts.items(), columns=["Type", "Count"])
df = df.sort_values("Count", ascending=False).reset_index(drop=True)
display(df)
```

## Close

```python
repo.close()
print("Driver closed")
```
